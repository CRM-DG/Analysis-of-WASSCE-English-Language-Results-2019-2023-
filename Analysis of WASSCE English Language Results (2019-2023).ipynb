```python
df4 = pd.read_excel("C:\\Users\\rdeni\\Desktop\\Mic.xlsx", sheet_name = 'Sheet5')
```

```python
df4
```

```python
# Create a mirrored pyramid plot
plt.figure(figsize=(10, 6))

plt.barh(df4['SUBJECT'], df4['SAT as % of entry'], color='green', label='Present')
plt.barh(df4['SUBJECT'], -df4['ABSENT AS % OF ENTRY'], color='red', label='Absent')

plt.xlabel('Number of Pupils')
plt.title('Mirrored Pyramid Plot for Candidates Present and Absent')
plt.legend()
plt.show()
```

```python
# Calculate the percentage of present and absent candidates for each column
for column in ['SAT as % of entry', 'ABSENT AS % OF ENTRY']:
    total = df4[column].sum()
    df4[f'{column} - Present %'] = (df4[column] / total) * 100
    df4[f'{column} - Absent %'] = 100 - df4[f'{column} - Present %']

# Create a bar chart with percentages
plt.figure(figsize=(12, 8))

bar_width = 0.4

plt.bar(df4['SUBJECT'], df4['SAT as % of entry - Present %'], width=bar_width, color='green', label='Present')
plt.bar(df4['SUBJECT'], -df4['ABSENT AS % OF ENTRY - Absent %'], width=bar_width, color='red', label='Absent')

# Annotate the bars with percentage values
for i, (present, absent) in enumerate(zip(df4['SAT as % of entry - Present %'], df4['ABSENT AS % OF ENTRY - Absent %'])):
    plt.text(i - bar_width / 2, present + 1, f'{int(present)}%', ha='center', va='bottom', color='black')
    plt.text(i + bar_width / 2, -absent - 1, f'{int(absent)}%', ha='center', va='top', color='black')

plt.xlabel('Subjects')
plt.ylabel('Percentage of Pupils')
plt.title('Bar Chart for Candidates Present and Absent')
plt.legend()
plt.show()
```

```python
# Calculate the percentage of present and absent candidates for each column
for column in ['SAT as % of entry', 'ABSENT AS % OF ENTRY']:
    total = df4[column].sum()
    df4[f'{column} - Present %'] = (df4[column] / total) * 100
    df4[f'{column} - Absent %'] = 100 - df4[f'{column} - Present %']

# Create a bar chart with percentages
plt.figure(figsize=(12, 8))

bar_width = 0.4

bars_present = plt.bar(df4['SUBJECT'], df4['SAT as % of entry'], width=bar_width, color='green', label='Present')
bars_absent = plt.bar(df4['SUBJECT'], -df4['ABSENT AS % OF ENTRY'], width=bar_width, color='red', label='Absent')

# Annotate the bars with percentage values
for bar, present, absent in zip(bars_present, df4['SAT as % of entry - Present %'], df4['ABSENT AS % OF ENTRY - Absent %']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{int(present)}%', ha='center', va='bottom', color='black')

for bar, absent in zip(bars_absent, df4['ABSENT AS % OF ENTRY - Absent %']):
    plt.text(bar.get_x() + bar.get_width() / 2, -bar.get_height(), f'{int(absent)}%', ha='center', va='top', color='black')

plt.xlabel('Subjects')
plt.ylabel('Number of Pupils')
plt.title('Bar Chart for Candidates Present and Absent')
plt.legend()
plt.show()
```
